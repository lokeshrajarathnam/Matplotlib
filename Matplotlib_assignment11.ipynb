```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as mplt
```

```python
url='https://raw.githubusercontent.com/Geoyi/Cleaning-Titanic-Data/master/titanic_original.csv'
titanic = pd.read_csv(url)
```

```python
titanic.columns
```

```python
titanic.head(5).shape
```

```python
male_sum = titanic[titanic.sex=='male']['sex'].value_counts()
female_sum = titanic[titanic.sex=='female']['sex'].value_counts()
```

```python
male_sum
```

```python
female_sum
```

```python
#1. Create a pie chart presenting the male/female proportion

mplt.pie([male_sum,female_sum],labels=['MALE','FEMALE'],autopct='%1.1f%%')
```

```python

mplt.show()
```

```python
#2. Create a scatterplot with the Fare paid and the Age, differ the plot color by gender
a = titanic['fare']
b= titanic['age']
c= titanic['sex']
colors = np.where(titanic['sex'] =='female','r','g')
scatter_plot  = mplt.scatter(a,b,alpha=0.5,c=colors)
mplt.xlabel('Fare')
mplt.ylabel('Age')
mplt.show()
```
